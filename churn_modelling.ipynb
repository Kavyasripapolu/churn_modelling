```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
data = pd.read_csv(r"C:\Users\mohan rao papolu\OneDrive\chinni\Datasets\Churn_Modelling.csv")
```

```python
x=data.drop(columns=['RowNumber','CustomerId','Surname','Exited'])
y=data['Exited']
y
```

```python
x
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
x=pd.get_dummies(x,columns=['Geography' ,'Gender'],drop_first=True)	
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
print("x train: ",x_train.shape)
print("x test: ",x_test.shape)
print("y train: ",y_train.shape)
print("y test: ",y_test.shape)
```

1. Linear Regression

```python
model=LinearRegression()
model.fit(x_train, y_train)
y_pred= model.predict(x_test)
```

```python
y_pred
```

2. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model=LogisticRegression(max_iter= 7000, random_state=42)
log_model.fit(x_train, y_train)
log_pred= log_model.predict(x_test)
```

```python
log_pred
```

```python
accuracy= accuracy_score(y_test ,log_pred)
report= classification_report(y_test, log_pred)
conf_matrix=confusion_matrix(y_test,log_pred)
```

```python
print(f'accuracy: {accuracy}')
print(f'report: {report}')
print(f'conf_matrix: {conf_matrix}')
```

3. Desicion Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
des_tree=DecisionTreeClassifier(random_state=42)
des_tree.fit(x_train,y_train)
```

```python
tree_pred=des_tree.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,tree_pred)
print(f'Accuracy Score: {accuracy:.4f}')
```

4. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randomforest=RandomForestClassifier(random_state=42,n_estimators=100,max_depth=10)
randomforest.fit(x_train, y_train)
```

```python
ran_pred=randomforest.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,ran_pred)
print(f'Accuracy Score: {accuracy:.4f}')
```

5. Ada boost

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
Adaboost=AdaBoostClassifier(n_estimators=50,learning_rate=1,random_state=42)
Adaboost.fit(x_train, y_train)
```

```python
Ada_pred=Adaboost.predict(x_test)
accuracy=accuracy_score(y_test, Ada_pred)
print(f'Accuracy Score: {accuracy:.4f}')
```

7. K-Nearest Neighbor(KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)  
knn_model.fit(x_train, y_train)
```
